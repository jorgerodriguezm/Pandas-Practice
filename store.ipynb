```python
%matplotlib inline
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_csv('superstore_data.csv')
data.head()
```

```python
data.dtypes
```

```python
data[['Education','Marital_Status']]=data[['Education','Marital_Status']].astype('string')
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'])
```

```python
data.dtypes
```

```python
data.head()
```

```python
data.columns
```

```python
data.Education.value_counts()
```

```python
columns=['Id', 'Year_Birth', 'Education', 'Marital_Status', 'Income', 'Kidhome',
       'Teenhome', 'Dt_Customer', 'Recency', 'MntWines','MntMeatProducts','NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
       'Response', 'Complain']
```

```python
data=data.loc[:,columns]
data.head(n=10)
```

```python
data['Age']=2022-data['Year_Birth']
data.head()
```

```python
data['Patronage']=(pd.Timestamp('2022-01-01')-data['Dt_Customer']).astype('timedelta64[Y]')
data.head()
```

```python
data.columns
```

```python
data=data[['Id','Year_Birth', 'Age','Dt_Customer','Patronage','Education', 'Marital_Status', 'Income', 'Kidhome',
       'Teenhome','Recency', 'MntWines', 'MntMeatProducts','NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases', 'NumWebVisitsMonth', 'Response', 'Complain']]
data.head(n=10)
```

```python
data['Patronage']=data['Patronage'].astype(int)
data.dtypes
```

```python
data=data[data['Age']<100]
data.head()
```

```python
data.Age.value_counts()
```

```python
purchases=data[['Marital_Status','NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases', 'NumWebVisitsMonth','Response']].groupby('Marital_Status').sum()
purchases
```

```python
eds_complain=data[['Education','Complain']].groupby('Education').sum()
eds_complain
```

```python
marital_complain=data[['Marital_Status','Complain']].groupby('Marital_Status').sum()
marital_complain
```

```python
data[['Education','MntWines','MntMeatProducts']].groupby('Education').sum()
```

```python
age_fig=px.histogram(data,x='Age', title='Age Distribution', color='Education')
age_fig.update_layout(bargap=0.2)
age_fig.show()
```

```python
income_fig=px.histogram(data,x='Income',color='Education', title='Income Distribution')
income_fig.update_layout(bargap=0.2)
income_fig.show()
```

```python
pat_fig=px.histogram(data,x='Patronage', title='Years of Patronage', text_auto=True)
pat_fig.update_layout(bargap=0.2)
pat_fig.show()
```
